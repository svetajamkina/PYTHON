Unsupervised training examples(clustering)

https://scikit-learn.org/dev/api/sklearn.datasets.html

```python
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(n_samples=1000, centers=6, cluster_std=0.80)
```

```python
X[:,0] #x coordinate
X[:,1] # y coodr
y #the cluster of the point
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1], c=y)
```

```python
plt.scatter(X[:,0],X[:,1], c='grey',edgecolors='k')
```

```python
from sklearn.cluster import KMeans # import model
```

```python
kmeans =KMeans(n_clusters=2) #initialize the model
```

```python
kmeans.fit(X) #train the model
```

```python
kmeans.predict(X) #predict
```

```python
cluster_labels = kmeans.predict(X) #predict
```

```python
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels, edgecolors='k')
```

```python
#get cluster centers
kmeans.cluster_centers_[:,0]
kmeans.cluster_centers_[:,1]
```

```python
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels, edgecolors='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, marker='*', c='red')
```

```python
kmeans =KMeans(n_clusters=3)
```

```python
kmeans.fit(X)
```

```python
kmeans.predict(X)
```

```python
cluster_labels = kmeans.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels, edgecolors='k')
```

```python
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels, edgecolors='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, marker='*', c='red')
```
