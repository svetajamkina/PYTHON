MATPLOT lib

```python
#Witn sysntetic data
# Sample data representing the average temperature (in degrees Celsius) from 2010 to 2014
data = {'Year': [2010, 2011, 2012, 2013, 2014],
        'Avg_Temperature': [14.1, 14.3, 14.5, 14.2, 14.6]}
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(data)
df
```

```python
#Version one
plt.plot(df['Year'], df['Avg_Temperature'],marker='o') #first value is x, second is y
plt.title('Average Temperature from 2010 to 2014') # we added a title
plt.xlabel('Year') # we added a label to the x axis
plt.ylabel('Temperature_in_Celsius') # we added a label to the y axis

plt.xticks(df['Year']) # we added a tick marks, year values are rounded
plt.grid(True) # we added a grid
plt.show()#show you a visualization
```

VERSION TWO

```python
# VERSION TWO

df.plot(x='Year', y='Avg_Temperature', kind='line')#we used the pandas library
plt.title('Average temperature from 2010 to 2014') #we added title
plt.xlabel('Year') #title on x axis
plt.ylabel('Temperature') #title on y axis

plt.xticks(df['Year']) #year values are rounded
plt.grid(True)
plt.show() #show you the visualization
```

# BAR CHART


```python
plt.bar(df['Year'], df['Avg_Temperature'], color='skyblue')
plt.title('Average temperature from 2010 to 2014')
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.xticks(df['Year'])
plt.grid(True)

```

```python
#H-BAR CHART
plt.barh(df['Year'], df['Avg_Temperature'], color='orange')
plt.title('Average temperature from 2010 to 2014')
plt.xlabel('Temperature')
plt.ylabel('Year')
```

#HISTOGRAMM-present the destribution of a single variable, frequency, numerical data

```python
data2 = {'Scores': [85, 92, 78, 20, 89, 90, 82, 56,60,23,90,40,77]}
df2 = pd.DataFrame(data2)
df2

plt.hist(df2['Scores'], bins=3,color='green', edgecolor='black')
```

```python
#CUSTOM BIN EDGES
bin_edges = [0, 20, 40, 60, 80, 100]
plt.hist(df2['Scores'], bins=bin_edges, color='green', edgecolor='black')
```

```python
# SCATTER PLOT -used to visialize the relationship between two variables

data3 = {'X': [1,2,3,5,5,2], 'Y':[5,8,7,4,9,5]}

df3 = pd.DataFrame(data3)

plt.scatter(x = df3['X'], y=df3['Y'])
plt.title('Scatter Plot')
plt.grid(True)
plt.show()
```

```python
# PIE CHART - represents parts of a whole

data4 = {'Category': ['A', 'B', 'C', 'D'], 'Values': [15, 30, 25, 40]}
df4 = pd.DataFrame(data4)
df4

plt.pie(df4['Values'], labels=df4['Category'], autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Pie Chart')
plt.show()
```

```python
# BOX PLOT - distribution of a dataset and identifies outliers
plt.boxplot(df4['Values'])
plt.grid(True)
plt.show()
```
