```python
 # the angles for the star points
angles = np.linspace(0, 2 * np.pi, 10, endpoint=False)

# radius for inner and outer points
outer_radius = 1
inner_radius = 0.5

# coordinates
x = np.array([outer_radius * np.cos(angles[i]) if i % 2 == 0 else inner_radius * np.cos(angles[i])
               for i in range(10)] + [outer_radius * np.cos(angles[0])])  # Close the star

y = np.array([outer_radius * np.sin(angles[i]) if i % 2 == 0 else inner_radius * np.sin(angles[i])
               for i in range(10)] + [outer_radius * np.sin(angles[0])])  # Close the star

# Plot the star outline
plt.figure(figsize=(6, 6))
plt.plot(x, y, color='gold', linewidth=2)  # Only plot the outline
plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)
plt.title('Star Shape')
plt.gca().set_aspect('equal')  # Equal aspect ratio
plt.grid(False)  # Turn off the grid
plt.show()
```

#Random walk

```python
n= 100000 #defining number of steps

x=np.zeros(n)
y=np.zeros(n)
```

```python
import random
val = random.randint(1, 4) # to get a direction of the step (4 directions in total)
val
```

```python
for i in range(1, n):
    val = random.randint(1, 4) # to get a direction of the step (4 directions in total)

    if val == 1: # dot moves x coordinate +1
        x[i] = x[i - 1] + 1
        y[i] = y[i - 1]

    elif val == 2: # dot moves x coordinate -1
        x[i] = x[i - 1] - 1
        y[i] = y[i - 1]

    elif val == 3: # dot moves y coordinate +1
        x[i] = x[i - 1]
        y[i] = y[i - 1] + 1

    else: # dot moves y coordinate -1
        x[i] = x[i - 1]
        y[i] = y[i - 1] - 1
```

```python
import pylab
```

```python
pylab.title('Random walk')
pylab.plot(x, y)
pylab
```

```python
plt.figure(figsize=(10, 6))  # Increase figure size
plt.plot(x, y, color='blue', lw=2, alpha=0.7)  # Line with transparency
plt.scatter(x[0], y[0], c='green', s=100, label='Start')  # Mark start point
plt.scatter(x[-1], y[-1], c='red', s=100, label='End')  # Mark end point

plt.title('2D Random Walk', fontsize=16)
plt.xlabel('X-Coordinate')
plt.ylabel('Y-Coordinate')
plt.grid(True)
plt.legend()

plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
steps = 100000
x = np.cumsum(np.random.choice([-1, 1], size=steps))
y = np.cumsum(np.random.choice([-1, 1], size=steps))

plt.figure(figsize=(10, 6))

# creates colors
colors = plt.cm.viridis(np.linspace(0, 1, len(x) - 1))

for i in range(len(x) - 1):
    plt.plot(x[i:i+2], y[i:i+2], color=colors[i], lw=2, alpha=0.7)

plt.scatter(x[0], y[0], c='green', s=100, label='Start')
plt.scatter(x[-1], y[-1], c='red', s=100, label='End')

plt.title('2D Random Walk', fontsize=16)
plt.xlabel('X-Coordinate')
plt.ylabel('Y-Coordinate')
plt.grid(True)
plt.legend()

plt.show()
```

```python
import random

def random_walk_1d(steps):
    position = 0
    up_steps = 0

    for _ in range(steps):
        step = random.choice([-1, 1])
        position += step
        if step == 1:
            up_steps += 1

    return position, up_steps

total_steps = 1000
final_position, total_up_steps = random_walk_1d(total_steps)

print(f"Final Position: {final_position}")
print(f"Total Up Steps: {total_up_steps} out of {total_steps}")
```
