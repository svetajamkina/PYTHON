```python
import numpy as np
```

```python
coin= ["Heads", "Tails"] #one dimentional array==list
```

```python
coin
```

```python
number_of_throws=1000 #size
```

```python
tosses= np.random.choice(coin, number_of_throws)
```

```python
print(f"The results of coin toss simulation for {number_of_throws} throws are:{tosses}")
```

```python
#Number of Heads
nr_heads = np.sum(tosses=="Heads")
```

```python
#Number of Tails
nr_tails = np.sum(tosses=="Tails")
```

```python
#Probabilities
prob_heads = nr_heads/number_of_throws
prob_tails = nr_tails/number_of_throws
```

```python
print(f"Probability of Heads: {prob_heads}")
print(f"Probability of Tails: {prob_tails}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 100000 random coordinates
coordinates3 = np.random.randn(100000, 2)

x = coordinates3[:, 0] # X values
y = coordinates3[:, 1] # Y values

# Create a color range based on the y-values or any other function of coordinates
colors = np.arctan2(y, x)  # This will create a circular color gradient

# Create scatter plot with rainbow colors
plt.scatter(x, y, c=colors, cmap='rainbow', alpha=0.7)
plt.title('100000 randomly selected coordinates - normal distribution')
plt.xlabel('X')
plt.ylabel('Y')
plt.colorbar()  # Adds a color bar for reference
plt.grid(True)
plt.show()
```

```python
coin = ["H", "T"]
number_of_tosses = 1000

# Toss two coins
coin1 = np.random.choice(coin, number_of_tosses, replace=True)
coin2 = np.random.choice(coin, number_of_tosses, replace=True)

# Combine the results
outcomes = np.char.add(coin1, coin2)

# Calculate each outcome
hh_count = np.sum(outcomes == "HH")
ht_count = np.sum(outcomes == "HT")
th_count = np.sum(outcomes == "TH")
tt_count = np.sum(outcomes == "TT")
outcome_counts = {"HH": hh_count, "HT": ht_count, "TH": th_count, "TT": tt_count}

print("***************************************************************************")
# Print the total counts
for outcome, count in outcome_counts.items():
    print(f"Total count for {outcome}: {count}")

# Calculate probability
probabilities = {outcome: count / number_of_tosses for outcome, count in outcome_counts.items()}

print("***************************************************************************")
# Print probabilities
for outcome, probability in probabilities.items():
    print(f"Probability of {outcome}: {probability:.4f}, when theoretical probability is 0,25.")

print("***************************************************************************")

# Plotting  bar chart
plt.bar(outcome_counts.keys(), outcome_counts.values(), color=['lightblue', 'lightgreen', 'lightcoral', 'peachpuff'], edgecolor='black', linewidth=0.5)
plt.xlabel('Outcomes')
plt.ylabel('Frequency')
plt.title('Frequency of Outcomes for Two Coin Tosses 1000# tosses')
plt.show()
```
