#HISTOGRAM


```python
# analyze Total for each pokemon
plt.hist(pokemon_df['Total'], color='green', edgecolor='black', alpha=0.7) #alpha = transparency, default = 10bins

plt.xlabel('Total')
plt.ylabel('Number of pokemons (Frequency)')

#avg line
plt.axvline(pokemon_df['Total'].mean(), linestyle='dashdot', color='grey')

#min line
plt.axvline(pokemon_df['Total'].min(), linestyle='dashdot', color='red')

#max line
plt.axvline(pokemon_df['Total'].max(), linestyle='dashdot', color='blue')

plt.grid(linestyle='--', alpha=0.7)
```

```python
pokemon_df['Total'].mean()
```

```python
value = plt.hist(pokemon_df['Total'], bins = 3, color='green', edgecolor='black', alpha=0.7)

hist_values = value[0]
bin_edges = value[1]

print('Histogram counts:', hist_values)
print('Bin edges', bin_edges)
```

```python
#Answer the following questions:
#1. How many Pokmons are with 'Type 1' == Water as a % of total?
#2. What is the maximum 'Speed' value? What is the minimum 'Speed' value? What is the difference between max and min 'Speed'?
#3. Filter the DataFrame to include only the Pokmon with 'Speed' >= 80.
#How many Pokmon meet this criterion? Display this DataFrame using your preferred visualization method.
#4. (DIFFICULT) Find Pokmon with the longest name (excluding spaces)? What is this pokemons name?
```

```python
#1
water_pokemon_percentage = (pokemon_df['Type 1'] == 'Water').mean() * 100
print(f"Percentage of Water-type Pokmons: {water_pokemon_percentage:.2f}%")
```

```python
#2
max_speed = pokemon_df['Speed'].max()
min_speed = pokemon_df['Speed'].min()
speed_difference = max_speed - min_speed
print(f"Maximum Speed: {max_speed}")
print(f"Minimum Speed: {min_speed}")
print(f"Difference between max and min Speed: {speed_difference}")
```

```python
#3
high_speed_pokemon = pokemon_df[pokemon_df['Speed'] >= 80]
num_high_speed_pokemon = len(high_speed_pokemon)
print(f"Number of Pokmon with Speed >= 80: {num_high_speed_pokemon}")
```

```python
plt.hist(high_speed_pokemon['Speed'], bins=10, color='green', edgecolor='black', alpha=0.7)
plt.xlabel('Speed')
plt.ylabel('Number of Pokmons')
plt.title('Distribution of Pokmon with Speed >= 80')
```

```python
plt.hist(pokemon_df['Speed'], color='green', edgecolor='black', alpha=0.7) #alpha = transparency, default = 10 bins
plt.xlabel('Speed')
plt.ylabel('Number of Pokemons(Frequency)')

plt.axvline(pokemon_df['Speed'].mean(), color='red', linestyle='dashed', linewidth=2)
plt.title('Average Speed')

plt.axvline(pokemon_df['Speed'].max(), color='blue', linestyle='dashed', linewidth=2)
plt.axvline(pokemon_df['Speed'].min(), color='orange', linestyle='dashed', linewidth=2)

plt.grid(linestyle='--', alpha=0.7)
plt.show()
```

```python
#4
pokemon_df['NameLengthNoSpaces'] = pokemon_df['Name'].str.replace(" ", "").str.len()

longest_pokemon_name_idx = pokemon_df['NameLengthNoSpaces'].idxmax()

longest_pokemon_name = pokemon_df.iloc[longest_pokemon_name_idx]["Name"]

print(f'The longest Pokmon name is {longest_pokemon_name}.')
```
