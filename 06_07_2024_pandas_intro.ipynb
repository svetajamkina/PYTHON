```python
import pandas as pd
```

```python
data_list = [1,2,3,4,5] #list
print(data_list)
```

Pandas series

```python
series_from_list = pd.Series(data_list) #a huge obs at the same time, 2 million number in 1 column, first index, 2-column
print(series_from_list)
```

Panda DataFrames


```python
dataframe_from_list = pd.DataFrame(data_list) # dataframe has two coordinates x and y, Have column name
print(dataframe_from_list)
```

Pandas Series from dictionary

```python
data_dict = {'A': 10, 'B': 20, 'C': 30, 'D': 40, 'E': 50} # dictionary object key-value paris

dataframe_from_dict = pd.Series(data_dict)
print(dataframe_from_dict)
#dtype: int64-integer values
```
